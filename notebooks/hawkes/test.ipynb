```python
from fim.data_generation.grid_samplers import sample_kernel_grid


# Example usage
rate = 0.0001  # Rate parameter (lambda)
max_value = 600000.0  # Maximum value for the range
size = (1, 3, 100)  # Shape of the output tensor

samples = sample_kernel_grid("exponential", rate=rate, max_value=max_value, size=size)
```

```python
import matplotlib.pyplot as plt


# Plot the samples
plt.hist(samples[0, 0, :].numpy(), bins=30, edgecolor="black")
plt.title("Histogram of Exponential Samples")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```
